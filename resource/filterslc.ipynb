```python
import pandas as pd
```

```python
month_data = []
for i in range(1, 13):
    filename = 'On_Time_On_Time_Performance_2017_' + str(i) + '.csv'
    data = pd.read_csv(filename)
    month_data.append(data)
```

```python
data = pd.concat(month_data)
```

```python
data
```

```python
toslc = data[data['Dest'] == 'SLC']
toslc = toslc.reset_index(drop=True)
toslc.to_csv('toslc2017.csv')
```

```python
fromslc = data[data['Origin'] == 'SLC']
fromslc = fromslc.reset_index(drop=True)
fromslc.to_csv('fromslc2017.csv')
```
